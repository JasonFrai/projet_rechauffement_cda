```python
import pandas as pd

# Affichage du dataset relatif aux GES
df_ges = pd.read_csv('EMISSIONS_ANNUAL_1830-2021.csv')
display(df_ges.head())
```

```python
# Units 
# 1 Tg = 1 million tons
# 1 Pg = 1 billion tons = 1000 millions tons

# Mise à l'echelle - Application d'un facteur 1000 aux données relatives au CO2

df_ges.loc[df_ges['Gas'] == 'CO[2]', 'Data'] = df_ges.loc[df_ges['Gas'] == 'CO[2]', 'Data'].apply(lambda x: x*1000)
df_ges.head()
```

```python
# Creation d'un DataFrame df_global (données mondiales)
df_ges_global = df_ges.loc[df_ges['CNTR_NAME'] == 'GLOBAL']
display(df_ges_global)
```

```python
df_ges_final = df_ges_global.groupby(['Gas', 'Year']).sum()
df_ges_final

reset_Gas = df_ges_final.reset_index(level='Gas')
reset_Gas

df_ges_final_p = reset_Gas.pivot(columns='Gas', values='Data')
df_ges_final_p
```

```python
# Affichage du dataset de la NASA
df_temp = pd.read_csv('ZonAnn.Ts+dSST_Zonal_annual_means.csv')
display(df_temp.head())
```

```python
# Concaténation

df = df_temp.merge(df_ges_final_p, how ='inner', on = 'Year')
df
```
