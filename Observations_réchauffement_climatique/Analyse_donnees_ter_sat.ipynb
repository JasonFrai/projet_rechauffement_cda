```python
import pandas as pd
import numpy as np

# Données satellitaires

df_sat = pd.read_csv('GLB.Ts+dSST_AIRS_Global.csv', header=1, skiprows = lambda x: x not in range(24, 48))
display(df_sat)

# Données terrestres

df_ter = pd.read_csv('GLB.Ts+dSST_AIRS_Global.csv', header=1, skiprows = lambda x: x not in range(48, 72))
display(df_ter)
```

```python
# Extraction des colonnes 'Year' et 'J-D'
df_sat_ext = df_sat[["Year","J-D"]]
df_ter_ext = df_ter[["Year","J-D"]]

# Concaténation
df = pd.merge(left = df_sat_ext, right = df_ter_ext, on = 'Year', how='inner')
df

# Suppression des lignes 2002 et 2023 (incompletes)
df.drop([0], axis=0, inplace=True)
df.drop([21], axis=0, inplace=True)

# Renommage des colonnes (J-D_x et J-D_y)
df.rename(columns = {'J-D_x': 'Anomalie_sat', 'J-D_y': 'Anomalie_ter'}, inplace = True) 
df
```

```python
# Correction du type des colonnes
print(df.dtypes)

df.Anomalie_sat = df.Anomalie_sat.astype(float)
df.Anomalie_ter = df.Anomalie_ter.astype(float)

print('\n')
print(df.dtypes)
```

```python
# Courbe

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(df["Year"], df["Anomalie_sat"], label="Anomalies satellitaires");
plt.plot(df["Year"], df["Anomalie_ter"], label="Anomalies terrestres");
plt.title('Anomalies annuelles satellitaires et terrestres')
plt.xlabel('Année')
plt.ylabel('Variation annuelle')
plt.xticks(np.arange(2003, 2022, step=3))
plt.legend()
plt.show()
```
