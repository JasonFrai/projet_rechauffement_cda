```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df2 = pd.read_csv('rechauffement.csv', header = 2)
df2.head()
```

```python
df2['Indicator Name'].value_counts()
```

```python
display(df2.info())
display(df2.isna().sum())
```

```python
df3 = df2[df2['Indicator Name'] == 'Hauteur moyenne des précipitations (mm par an)']
df3 = df3.drop(['Country Code','Indicator Name','Indicator Code'], axis = 1)
df3.head()
```

```python
df3.replace(' ', pd.NaT, inplace=True)
df3.dropna(axis = 1 , how = 'all', inplace = True)
df3.head()
```

```python
df3.reset_index(drop = True)
df3_t = df3.transpose().copy()
df3_t = df3_t.set_axis(df3_t.iloc[0], axis=1).drop(df3_t.index[0])
df3_t = df3_t.reset_index(drop = False)
df3_t = df3_t.rename(columns = {'index' : 'date','Country Name' : ' '})
df3_t.head()
```

```python
fig, ax = plt.subplots(figsize=(13, 8))

for y in df3_t.columns:
   
    y_values = df3_t[y].values.tolist()

    x = df3_t['date']

    ax.plot(x, y_values, label=y)

ax.legend()
ax.set_title("Emission total de GES depuis 1990 par pays")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xticks(rotation = 90)

plt.show()
```
