```python
import pandas as pd

df = pd.read_csv('antarctic-ozone-hole-area.csv')
df
```

```python
df['average'] = df['Mean ozone hole area'].rolling(window=5).mean()
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

ax = df.plot(x = 'Year', y = ['Mean ozone hole area', 'average'], 
        label=['Surface moyenne du trou', 'Moyenne glissante sur 5 ans'],
        title = "Surface du trou dans la couche d'ozone observ au dessus de l'Antarctique",
        figsize=(12,5))
ax.set_xlim([1979, 2022])
ax.set_yticks([5000000, 10000000, 15000000, 20000000, 25000000, 30000000])
ax.set_yticklabels(['5 millions km', '10 millions km','15 millions km', '20 millions km', '25 millions km', 
                    '30 millions km'])
ax.set(xlabel='Anne')
ax.legend(fontsize=10, loc='upper left');
```
