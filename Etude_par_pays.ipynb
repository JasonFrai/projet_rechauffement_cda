```python
cata_nat = pd.read_csv(r"C:\Users\jason\Documents\Project rechauffement\pend-gdis-1960-2018-disasterlocations.csv")
#Source = SEDAC
cata_nat.sort_values(by="year",ascending = True,inplace = True)
#cata_nat["id"].unique().shape
#cata_nat = cata_nat.set_index("year")
cata_nat
```

```python
#Suppression doublon
cata_nat2 = cata_nat.drop_duplicates(subset = "id")
#24N-90N	24S-24N	90S-24S	64N-90N	44N-64N	24N-44N	EQU-24N	24S-EQU	44S-24S	64S-44S	90S-64S
cata_nat2[cata_nat2["latitude"]<-44]
zone_geo = []
for element in cata_nat2['latitude'] :
    if element < -64 :
        zone_geo.append("90S-64S")
    elif element < -44 :
        zone_geo.append("64S-44S")
    elif element  < -24 :
        zone_geo.append("24S-44S")
    elif element < 0 :
        zone_geo.append ("EQU-24S") 
    elif element <24 :
        zone_geo.append ( "24N-EQU")
    elif element < 44 :
        zone_geo.append ( "44N-24N")
    elif element < 64 :
        zone_geo.append ("64N-24N")
    elif 1==1 : 
        zone_geo.append("90N-64N" )
cata_nat2["zone"] = zone_geo
cata_nat2
```

```python
final_cata = cata_nat2[['country','disastertype','zone','year']]
final_cata['year'] = final_cata['year'].astype(str)
final_cata['zone'].value_counts()
final_cata['zone'].value_counts().plot(kind='bar')
```
