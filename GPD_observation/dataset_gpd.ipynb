```python
rechauffement = pd.read_csv(r"C:\Users\jason\Downloads\ZonAnn.Ts+dSST (1).csv")
dico = {'Year':'index'}
rechauffement = rechauffement.rename(dico, axis = 1) 
rechauffement = rechauffement.set_index(['index'])
rechauffement
```

```python
rechauffement_T = rechauffement.T
rechauffement_T = rechauffement_T.reset_index()
rechauffement_T = rechauffement_T.add_suffix('_température')
dico = {'index_température':'lat'}
rechauffement_T = rechauffement_T.rename(dico, axis = 1) 
rechauffement_T
```

```python
latitude_pays = pd.read_csv(r"C:\Users\jason\Downloads\countries_codes_and_coordinates.csv")
latitude_pays['Latitude (average)'] =latitude_pays['Latitude (average)'].apply(lambda x : x[2:])
latitude_pays['Latitude (average)'] =latitude_pays['Latitude (average)'].apply(lambda x : x[:-1]) 
latitude_pays['Latitude (average)'] =latitude_pays['Latitude (average)'].astype(float)
latitude_pays
```

```python
latitude_pays2 = latitude_pays[['Country','Latitude (average)']]
```

```python
def lat_pays(latitude_du_pays) :
    if latitude_du_pays <-64 :
        return '90S-64S'
    elif latitude_du_pays <-44 :
        return '64S-44S'
    elif latitude_du_pays <-24 :
        return '44S-24S'
    elif latitude_du_pays <-0 :
        return '24S-EQU'
    elif latitude_du_pays <24 :
        return 'EQU-24N'
    elif latitude_du_pays <44 :
        return '24N-44N'
    elif latitude_du_pays <64 :
        return '44N-64N'
    elif latitude_du_pays <90 :
        return '64N-90N'
```

```python
latitude_pays2['name'] = latitude_pays2.apply(lambda row: lat_pays(row['Latitude (average)']), axis=1)
```

```python
dico = {'name':'lat'}
latitude_pays2 = latitude_pays2.rename(dico, axis = 1) 
latitude_pays2
```

```python
new_df = latitude_pays2.merge(right = rechauffement_T , on = 'lat' , how = 'left') 
new_df
```

```python
df_final = world_map.merge(right = new_df , on = 'Country' , how = 'left')
df_final
```

```python
ax = df_final.plot(figsize=(20, 8))
df_final.plot(ax = ax,cmap = 'jet',column = '2000_température', edgecolor = 'black',legend = True)
```

```python
cata_nat = pd.read_csv(r"C:\Users\jason\Documents\Project rechauffement\pend-gdis-1960-2018-disasterlocations.csv")
display(cata_nat["disastertype"].unique())
cata_nat = cata_nat[cata_nat["disastertype"]!='earthquake']
cata_nat = cata_nat[cata_nat["disastertype"]!='volcanic activity']
dico = {'country':'Country'}
cata_nat = cata_nat.rename(dico, axis = 1) 
cata_nat
```

```python
pays = []
disaster = []

for country in cata_nat['Country'].unique() :

    pays.append(country)
    disaster.append(cata_nat[cata_nat['Country']==country].shape[0])      

df_cata =  pd.DataFrame(zip(pays,disaster),columns=['Country', 'catastrophe'])

for anne in cata_nat['year'].unique() :
    
  df_cata = df_cata.merge(right = cata_nat[cata_nat['year']==anne].Country.value_counts().reset_index(name=anne) , on = 'Country' , how = 'left')
df_cata = df_cata.fillna(0)
df2 = df_cata[['Country','catastrophe']]
df_cata.drop(['catastrophe','Country'],axis = 1,inplace = True)
df_cata = df_cata.add_suffix('_catastrophe')
df_cata[['Country','catastrophe']] = df2[['Country','catastrophe']] 
df_cata.head()
```

```python
df_final = df_final.merge(right = df_cata , on = 'Country' , how = 'left')
df_final
```

```python
ax = df_final.plot(figsize=(20, 8))
df_final.plot(ax = ax,cmap = 'summer',column = 'catastrophe', edgecolor = 'black',legend=True)
```
